# Perceptron for AND Gate

## Objective
To design and train a single-layer perceptron to implement the logical AND gate.

## Description
The AND gate outputs 1 only when both inputs are 1.
A perceptron can learn AND because it is a linearly separable problem.


```python
import numpy as np

# AND gate dataset
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([0, 0, 0, 1])

# Initialize weights and bias
w = np.zeros(2)
b = 0
lr = 0.1

# Training
for _ in range(10):
    for i in range(len(X)):
        y_pred = 1 if np.dot(X[i], w) + b >= 0 else 0
        w += lr * (y[i] - y_pred) * X[i]
        b += lr * (y[i] - y_pred)

# Testing
print("AND Gate Output:")
for x in X:
    print(x, "->", 1 if np.dot(x, w) + b >= 0 else 0)
```
