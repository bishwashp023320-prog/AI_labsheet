# Perceptron for NOT Gate

## Objective
To design and train a perceptron to implement the NOT gate.


```python
import numpy as np

X = np.array([[0], [1]])
y = np.array([1, 0])

w = 0
b = 0
lr = 0.1

for _ in range(10):
    for i in range(len(X)):
        y_pred = 1 if X[i]*w + b >= 0 else 0
        w += lr * (y[i] - y_pred) * X[i]
        b += lr * (y[i] - y_pred)

print("NOT Gate Output:")
for x in X:
    print(x[0], "->", 1 if x*w + b >= 0 else 0)
```
