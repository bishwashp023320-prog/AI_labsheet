# Perceptron for XOR Gate

## Objective
To demonstrate that a single-layer perceptron cannot learn the XOR gate.

## Explanation
XOR is not linearly separable, hence cannot be solved using a single perceptron.


```python
import numpy as np

X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([0, 1, 1, 0])

w = np.zeros(2)
b = 0
lr = 0.1

for _ in range(10):
    for i in range(len(X)):
        y_pred = 1 if np.dot(X[i], w) + b >= 0 else 0
        w += lr * (y[i] - y_pred) * X[i]
        b += lr * (y[i] - y_pred)

print("XOR Gate Output (Fails):")
for x in X:
    print(x, "->", 1 if np.dot(x, w) + b >= 0 else 0)
```
